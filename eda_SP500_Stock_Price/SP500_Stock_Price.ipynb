```python
import pandas as pd
def load_data():
    url = "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
    html = pd.read_html(url, header=0)
    df = html[0]
    return df
```

```python
df = load_data()
df
```

```python
sectors_unique = df['GICS Sector'].unique()
sectors_unique
```

```python
df_selected_sector = df[ (df['GICS Sector'].isin(sectors_unique))]
df_selected_sector
```
