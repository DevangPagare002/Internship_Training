```python
import yfinance as yf
import streamlit as st
import pandas as pd
```

```python
st.write("""
# Simple Stock Price App

Shown are the stock **closing prices** and **volumes** of Google""")
```

```python
tickerSymbol = "GOOGL"
```

```python
tickerData = yf.Ticker(tickerSymbol)
```

```python
tickerDf = tickerData.history(period='1d', start='2010-05-31', end='2020-05-31')
```

```python
tickerDf.head()
```

```python
import matplotlib.pyplot as plt

plt.plot(tickerDf["Open"])
plt.ylabel("Opening price")
plt.show()
```

```python
plt.plot(tickerDf["Close"])
plt.ylabel("Closing Price")
plt.show()
```

```python
plt.plot(tickerDf["Volume"])
plt.ylabel("Volumes variations")
plt.show()
```
