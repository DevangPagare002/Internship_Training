```python
selected_year = 2019
url = 'http://www.pro-football-reference.com/years/' + str(selected_year) + '/rushing.htm'
```

```python
import pandas as pd
html = pd.read_html(url, header=0)
df = html[0]
# df.columns = range(df.shape[1])
# df.drop(df.columns)
df
```

```python
df.columns = df.iloc[0]
df
```

```python
df = df.drop(df[df.Age == 'Age'].index)
df
```

```python
df = df.fillna(0)
df
```

```python
df.columns
df.reset_index()
df = df.drop(['Rk'], axis=1)
```

```python
sorted_unique_pos = sorted(df['Pos'].unique())
sorted_unique_pos
```

```python
sorted_unique_teams = sorted(df['Tm'].unique())
sorted_unique_teams
```

```python
df_selected_team = df[(df.Tm.isin(sorted_unique_teams) & (df.Pos.isin(sorted_unique_pos)))]
```
